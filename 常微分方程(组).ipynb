# 常微分方程组求解

如果是一元的情况:不需要用到list,直接用一个变量存储即可,具体可见基础篇

```python

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

m=75
g=9.8
k=0.25

# 定义常微分方程组
def model(y, t):
    y1, y2 = y
    dydt = [-y1 + y2, -y1 - y2] # y1' = -y1 + y2, y2' = -y1 - y2
    return dydt

# 初始条件
y0 = [1, 0]

# 时间点
t = np.linspace(0, 10, 100)

# 求解ODE
solution = odeint(model, y0, t)

print(solution)

# 绘制结果
plt.plot(t, solution[:, 0], label='y1(t)')
plt.plot(t, solution[:, 1], label='y2(t)')
plt.xlabel('Time t')
plt.ylabel('Solutions y1, y2')
plt.legend()
plt.show()

```
