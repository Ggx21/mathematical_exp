```python
# 矩阵的条件数
def matrix_condition_number(A, p=None):
    return np.linalg.cond(A,p=p)

x=np.linspace(0, 2, 1000)
conds=[]
for i in range(len(x)):
  A=np.array([[x[i],4],[-1,x[i]]])
  cond=matrix_condition_number(A,1)
  conds.append(cond)
print("矩阵的条件数：", conds)

import matplotlib.pyplot as plt
plt.plot(x, conds)
plt.xlabel('x')
plt.ylabel('condition number')
plt.title('Condition number of matrix')
plt.show()
```
