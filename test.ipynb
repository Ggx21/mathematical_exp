```python
import matplotlib.pyplot as plt

# 数据
population_size = [10, 20, 30, 40, 50]
iteration_count = [20, 40, 60, 80, 100]
instruction_length = [10, 20, 30, 40, 50]

temperatures = {
    "Population Size": [56.382, 57.458, 57.996, 57.996, 59.072],
    "Iteration Count": [57.458, 59.072, 58.534, 60.148, 59.61],
    "Instruction Length": [59.61, 57.458, 57.996, 57.996, 57.996],
}

# 创建一个图形窗口，并设置1行3列的子图布局
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# 第一个子图：种群数量与温度的关系
axs[0].plot(
    population_size,
    temperatures["Population Size"],
    marker="o",
    linestyle="-",
    color="b",
)
axs[0].set_title("Population Size vs. Temperature")
axs[0].set_xlabel("Population Size")
axs[0].set_ylabel("Temperature")

# 第二个子图：迭代次数与温度的关系
axs[1].plot(iteration_count, temperatures["Iteration Count"], marker="s", linestyle="--", color="r")
axs[1].set_title("Iteration Count vs. Temperature")
axs[1].set_xlabel("Iteration Count")
axs[1].set_ylabel("Temperature")

# 第三个子图：指令长度与温度的关系
axs[2].plot(instruction_length, temperatures["Instruction Length"], marker="^", linestyle="-.", color="g")
axs[2].set_title("Instruction Length vs. Temperature")
axs[2].set_xlabel("Instruction Length")
axs[2].set_ylabel("Temperature")

# 调整子图间距
plt.tight_layout()

# 显示图形
plt.show()

# 保存图形
fig.savefig("temperature_relationships.png")
```
