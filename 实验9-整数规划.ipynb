# 整数规划

## 基本概念
实验 7 中我们已经给出了优化问题的数学规划模型的一般形式
$$
min z=f(x),x=(x_1,x_2,\cdots,x_n)\in R^n \tag{1}
$$
$$
s.t. g_i(x)\leq 0 , i=1,2,\cdots,m \tag{2}
$$
 
当模型中的决策变量x的分量$x_i,i=1,2,\cdots,n$至少有一个只取整数数值时,则该模型称为整数规划.(integer programming,IP)
- 当约束函数$g_i$和目标函数$f$都是决策变量的线性函数时,称为线性整数规划.(linear integer programming,LIP)
- 当所有决策变量都只能在整数集合中取值时,称为纯整数规划.(pure integer programming,PIP)
- 当部分决策变量只能在整数集合中取值时,称为混合整数规划.(mixed integer programming,MIP)
- 此外,当整数决策变量只能取0或1时,称为0-1整数规划.(0-1 integer programming)

## 整数规划的基本原理和方法

### 分支定界法

分支定界法是整数规划的一种常用方法,其基本思想是:将整数规划问题转化为一系列的线性规划问题,然后利用线性规划的方法求解这些线性规划问题,从而逐步逼近整数规划问题的最优解.

- 先找到线性松弛的解$x^0$,这里$x^0$未必是整数
- 如果求解的线性松弛问题的解$x^0$是整数,则$x^0$就是整数规划问题的最优解,否则就可以以$x^0$为基础,将整数规划问题分解为两个子问题,然后分别求解这两个子问题
- 两个子问题的解空间分别为$x_i\leq \lfloor x_i^0 \rfloor$和$x_i\geq \lfloor x_i^0 \rfloor+1$
- 然后再对两个子问题分别进行上述过程,若在某一时刻,得到了一个全整数解的目标值为$z_0$,则$z_0$为原解的一个上界
- 若从某个子问题开始求解到一个子问题的下界大于最好的一个$z_0=U$,则可以放弃这个子问题(称这个分支被杀死或已探明)
